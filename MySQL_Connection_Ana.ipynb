# Web-scrapping for additional information üî¶

```python
# Add the name and code of the country
import requests
from bs4 import BeautifulSoup
url = 'https://en.wikipedia.org/wiki/ISO_3166-1_alpha-3#ABW'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
soup.find_all(class_='monospaced')
soup.find_all(class_='monospaced')[366].get_text()
```

```python
# Extract the codes
country_code = []
i = 0
while i < 249:
    code = soup.find_all(class_='monospaced')[i].get_text()
    country_code.append(code)
    i += 1
```

```python
country_code[248]
#len(country_code)
```

```python
# Extract the names
country_name = []
i = 0
while i < 249:
    name = soup.find_all(class_='monospaced')[i].find_next().get_text()
    country_name.append(name)
    i +=1
```

```python
country_name[248]
#len(country_name)
```

```python
country_codes_df = pd.DataFrame({'code':country_code, 'country':country_name})
#country_codes_df
```

```python
# Store the country_codes as .csv for later use
country_codes_df.to_csv('country_codes_.csv', index = 1)
```

# Connecting to MySQL üîó

Now we can start the database in MySQL.
</br>The script looks like this:
</br>-- Create the initial Database for the WBS_DS_29 Project. It will contain five tables
</br>DROP DATABASE IF EXISTS sql_database_wbs_ds_29_project;
</br>CREATE DATABASE sql_database_wbs_ds_29_project;
</br>USE sql_database_wbs_ds_29_project;
</br>CREATE TABLE diseases (
</br>	disease_id INT AUTO_INCREMENT,
</br>    disease_name VARCHAR (100) NOT NULL,
</br>    disease_cause VARCHAR (100),
</br>    disease_transmission VARCHAR (100),
</br>    disease_incubation_period_min INT,
</br>    disease_incubation_period_min_unit VARCHAR (50),
</br>    disease_incubation_period_max INT,
</br>    disease_incubation_period_max_unit VARCHAR (50),
</br>    disease_contagious_period VARCHAR (50),
</br>    disease_contagious_period_unit VARCHAR (50),
</br>    PRIMARY KEY (disease_id)
</br>);
</br>CREATE TABLE diseases_stats_per_countries (
</br>	disease_stat_id INT AUTO_INCREMENT,
</br>    disease_stat_year INT NOT NULL,
</br>    disease_stat_cases INT,
</br>    country_id INT,
</br>    disease_id INT,
</br>    PRIMARY KEY(disease_stat_id),
</br>    FOREIGN KEY (country_id) REFERENCES countries(country_id),
</br>    FOREIGN KEY (disease_id) REFERENCES diseases(disease_id)
</br>);

As we can see, the *diseases_stats_df* has no unique diseases and has the statistical information. The *diseases* table in MySQL has unique values (i.e. Id) for each unique disease as well as information related to each disease. If we merge both tables we will have the information we need for the *diseases_stats_per_countries* table. The foreign key will then be the Id of the disease and later on from the country as well.  

```python
# Merge by the disease_name
diseases_stats_df = diseases_stats_df.merge(diseases_from_sql, on = 'disease_name', how = 'left')
```

```python
diseases_stats_df.info()
```
